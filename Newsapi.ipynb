```python
# Desenvolvido por Karla Zorzam em 07/03/2023
# Descrição: o código tem por objetivo automatizar a pesquisa no período de 25/02/2023 até a data presente
#           (07/03/2023) usando a NEWS API 
  

    
import requests
from pprint import pprint
import json



# objeto da pesquisa  
subject = 'startup in civil engineering'


# Chave API
key_file = open('newsapi_key.txt', 'r')
api_key = key_file.read()


url = (f'https://newsapi.org/v2/everything?'
       f'q={subject}&'
       'from=2023-02-25&'
       'to=2023-03-07&'
       'sortBy=popularity&'
       f'apiKey={api_key}')

resposta = requests.get(url)
#pprint(response.text)


dict_resposta = resposta.json() #transforma o texto em dicionário
lista_artigos = dict_resposta['articles'] #extrai a lista de artigos dentro do dicionário
print('A pesquisa exibe um total de {} resultados.'.format(len(lista_artigos)))
print('Segue abaixo a lista com os dados de autor, nome do artigo e descrição:\n')

for artigo in lista_artigos:
    #print(artigo)
    print('Author:', artigo['author'])
    print('Artigo:', artigo['title'])
    print('Descrição:',artigo['description'])
    print('-----------')

```
